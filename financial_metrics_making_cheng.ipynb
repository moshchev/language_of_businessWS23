```python
#1.transform 12 tables
#1.1 add company labels
# coke 10k
cola_balance.insert(0, 'Company Name', 'coca')
cola_income.insert(0, 'Company Name', 'coca')
cola_cf.insert(0, 'Company Name', 'coca')

# pepsi 10k
pepsi_balance.insert(0, 'Company Name', 'pepsi')
pepsi_income.insert(0, 'Company Name', 'pepsi')
pepsi_cf.insert(0, 'Company Name', 'pepsi')
```

```python
pepsi_balance
```

```python
pepsi_balance.columns = [col.strftime('%Y-%m-%d') if isinstance(col, pd.Timestamp) else col for col in pepsi_balance.columns]
pepsi_income.columns = [col.strftime('%Y-%m-%d') if isinstance(col, pd.Timestamp) else col for col in pepsi_income.columns]
pepsi_cf.columns = [col.strftime('%Y-%m-%d') if isinstance(col, pd.Timestamp) else col for col in pepsi_cf.columns]
```

```python
cola_balance.columns = [col.strftime('%Y-%m-%d') if isinstance(col, pd.Timestamp) else col for col in cola_balance.columns]
cola_income.columns = [col.strftime('%Y-%m-%d') if isinstance(col, pd.Timestamp) else col for col in cola_income.columns]
cola_cf.columns = [col.strftime('%Y-%m-%d') if isinstance(col, pd.Timestamp) else col for col in cola_cf.columns]
```

```python
#labelize data source
#cola 10k
cola_balance.insert(0, 'Data Source', 'bs')
cola_income.insert(0, 'Data Source', 'is')
cola_cf.insert(0, 'Data Source', 'cf')

# pepsi 10k
pepsi_balance.insert(0, 'Data Source', 'bs')
pepsi_income.insert(0, 'Data Source', 'is')
pepsi_cf.insert(0, 'Data Source', 'cf')
```

```python
# index to indicators
# cola
cola_balance = cola_balance.reset_index().rename(columns={'index': 'Financial Indicators'})
cola_income = cola_income.reset_index().rename(columns={'index': 'Financial Indicators'})
cola_cf = cola_cf.reset_index().rename(columns={'index': 'Financial Indicators'})

# pepsi
pepsi_balance = pepsi_balance.reset_index().rename(columns={'index': 'Financial Indicators'})
pepsi_income = pepsi_income.reset_index().rename(columns={'index': 'Financial Indicators'})
pepsi_cf = pepsi_cf.reset_index().rename(columns={'index': 'Financial Indicators'})
```

```python
pepsi_balance
```

```python
form10k_inone=pd.concat([cola_balance,cola_income,cola_cf,pepsi_balance,pepsi_income,pepsi_cf])
```
