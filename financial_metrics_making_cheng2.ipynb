```python
import pandas as pd

companies = form10k_inone['Company Name'].unique()

new_rows = []

for company in companies:
    company_data = form10k_inone[form10k_inone['Company Name'] == company]

    total_assets = company_data.iloc[:, 3:].loc[company_data['Financial Indicators'] == 'Total Assets'].values
    total_debt = company_data.iloc[:, 3:].loc[company_data['Financial Indicators'] == 'Total Debt'].values
    net_assets = total_assets - total_debt
    
    new_row_data = [net_assets[0, col - 3] for col in range(3, len(form10k_inone.columns))]
    new_row_data = ['Net Assets', 'bs', company] + new_row_data

    # create new rows to accommodate the new born metrics
    new_row = dict(zip(form10k_inone.columns, new_row_data))

    new_rows.append(new_row)

# insert the new records
form10k_inone = pd.concat([form10k_inone, pd.DataFrame(new_rows)], ignore_index=True)

# delete useless columns, if any
form10k_inone = form10k_inone.dropna(axis=1, how='all')



```

```python
import pandas as pd
import numpy as np

def avrg_balance(form10k_inone, financial_indicator):
    new_rows = []

    for _, company_data in form10k_inone.groupby('Company Name'):
        # get the data of the indicator so to calc average
        indicator_data = company_data.loc[company_data['Financial Indicators'] == financial_indicator, form10k_inone.columns[3:]]

        # calculate moving average
        avg_balances = [
            np.nanmean(indicator_data.iloc[:, col-2:col].values)
            for col in range(2, len(form10k_inone.columns))
        ]

        # create new row
        new_row_data = [financial_indicator + 'MA', 'bs', company_data['Company Name'].iloc[0]] + avg_balances

        # zip the row
        new_row = dict(zip(form10k_inone.columns, new_row_data))

        new_rows.append(new_row)

    # insert the new rows into the original df
    form10k_inone = pd.concat([form10k_inone, pd.DataFrame(new_rows)], ignore_index=True)

    # delete null columns, if any
    form10k_inone = form10k_inone.dropna(axis=1, how='all')

    return form10k_inone

# calc MA of Total Assets
form10k_inone = avrg_balance(form10k_inone, 'Total Assets')

```

```python
#calc MA of Accounts Payable
form10k_inone = avrg_balance(form10k_inone, 'Accounts Payable')
```

```python
#calc MA of Accounts Receivable
form10k_inone = avrg_balance(form10k_inone, 'Accounts Receivable')
```

```python
#calc MA of Inventory
form10k_inone = avrg_balance(form10k_inone, 'Inventory')
```

```python
# 1. Metrics 1 --making net income for profitability
corp = "coca"
nom = 'Net Income'
denom = 'Total Revenue'
nomsource = 'is'
denomsource = 'is'


def get_rate(corp, nom, denom, nomsource, denomsource):
    nom_data = form10k_inone[(form10k_inone['Data Source'].isin([nomsource])) & (form10k_inone['Company Name']== corp) & (form10k_inone['Financial Indicators'] == nom)]
    denom_data = form10k_inone[(form10k_inone['Data Source'].isin([denomsource])) & (form10k_inone['Company Name']== corp) & (form10k_inone['Financial Indicators'] == denom)]
    
    nom_values=nom_data.iloc[:,3:].values
    denom_values=denom_data.iloc[:,3:].values
    
    rate=nom_values/denom_values
    return rate

get_rate(corp,nom,denom,nomsource,denomsource)
```
